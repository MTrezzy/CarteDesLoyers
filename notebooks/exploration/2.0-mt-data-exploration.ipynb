## Understanding the region code

We can try to guess by looking at the cities that are in the region.

### Region 32

```python
mask = df_maisons_parquet.REG == 32
region32 = df_maisons_parquet[mask]
region32.shape
```

```python
region32_cities = region32['LIBGEO'].unique()
region32_cities
```

```python
region32_cities.size
```

```python
arr = np.arange(1, 3747, 300)
arr
```

```python
for i in arr:
    print(region32_cities[i])
```

By checking by hand some of the cities in the region with the code **32**, it is possible to conclude that the region 32 corresponds to the region **Hauts-de-France**.

### Region 84

```python
mask = df_maisons_parquet.REG == 84
region84 = df_maisons_parquet[mask]
region84.shape
```

```python
region84_cities = region84['LIBGEO'].unique()
region84_cities
```

The region **84** corresponds to **Auvergne-Rhône-Alpes**.
